```python
from IPython.display import FileLink
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, spearmanr
plt.rc('savefig', dpi=150)
DIR = '/media/raid/Zaloha/vypocty/mirek_Bb/lhs/'
freet_input = np.loadtxt(os.path.join(DIR, 'lhs.txt'))
```

```python
task_files = [v for v in os.listdir(DIR) if os.path.splitext(v)[1]=='.in']
task_files.sort()
```

```python
data = []
for i in task_files:
    filename = os.path.splitext(i)[0]
    datafile = os.path.join(DIR, 'results', filename + '.txt')
    if os.path.exists(datafile):
        d = np.loadtxt(datafile)
        data.append(d)
```

Analysis of peak values (force and CMOD)
=========================================

```python
peak_force = []
peak_cmod = []
for i in range(len(data)):
    force = data[i][:,3]
    cmod = data[i][:,6]-data[i][:,5]
    peak_force.append(force.max())
    peak_cmod.append(cmod[(force==force.max())[0]])
peak_cmod = np.array(peak_cmod)
peak_force = np.array(peak_force)
```

```python
print 'mean value', np.mean(peak_cmod)
print 'standard deviation', np.std(peak_cmod, ddof=1)
print 'skewness', skew(peak_cmod)
print 'kurtosis', kurtosis(peak_cmod)
plt.hist(peak_cmod, bins=50, normed=True)
plt.show()
```

```python
print 'mean value', np.mean(peak_force)
print 'standard deviation', np.std(peak_force, ddof=1)
print 'skewness', skew(peak_force)
print 'kurtosis', kurtosis(peak_force)
plt.hist(peak_force, bins=30, normed=True)
plt.show()
```

Per partes analysis
====================

```python
n = 3**np.arange(10)
idx_end = np.cumsum(n)
idx_start = idx_end - n
idx_start, idx_end
```

```python
mean_vals = []
std_vals = []
color = ['blue','green','red','black','cyan','magenta','#CCCCCC','#EEEEEE','red','red']
label = ['E', 'F_t', 'G_f', 'cmod', 'F']
for i in range(10):
    if corr_data.shape[0]>=idx_end[i]:
        m = np.mean(corr_data[idx_start[i]:idx_end[i], :], axis=0)
        s = np.std(corr_data[idx_start[i]:idx_end[i], :], axis=0)
        mean_vals.append(m)
        std_vals.append(s)
mean_vals = np.array(mean_vals)
std_vals = np.array(std_vals)

for i in range(mean_vals.shape[1]):
    plt.figure(0)
    plt.plot(mean_vals[:, i]/mean_vals[-1, i], '-x', color=color[i], label=label[i])
    #plt.plot((mean_vals[:, i]+std_vals[:, i])/mean_vals[:, i], '--x', color=color[i], linewidth=0.5)
    #plt.plot((mean_vals[:, i]-std_vals[:, i])/mean_vals[:, i], '--x', color=color[i], linewidth=0.5)
    #plt.yscale('log')
    plt.ylim(0.95,1.05)
    plt.legend(loc='best')
    
    plt.figure(1)
    plt.plot(np.arange(std_vals.shape[0]-1)+1, (std_vals[:, i]/std_vals[-1,i])[1:], '-x', color=color[i], label=label[i])
    plt.xlim(0,std_vals.shape[0]-1)
    plt.legend(loc='best')
    
#plt.figure(2)
#plt.plot(mean_vals/mean_vals[-1], '-x', color=color[i])
#plt.plot((mean_vals+std_vals)/mean_vals[-1], '--x', color=color[i], linewidth=0.5)
#plt.plot((mean_vals-std_vals)/mean_vals[-1], '--x', color=color[i], linewidth=0.5)
#plt.yscale('log')
#plt.ylim(0.5,2)

#plt.figure(1)
#plt.plot(std_vals/std_vals[-1], '-x', color=color[i])

plt.show()
```
