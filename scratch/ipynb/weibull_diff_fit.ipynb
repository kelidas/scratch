```python
from IPython.display import display, HTML
import os
import numpy as np
import mpmath as mp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from scipy import stats
import sympy as sp
#sp.init_session()
#x, y, z, t = sp.symbols('x y z t')
#k, m, n = sp.symbols('k m n', integer=True)
#f, g, h = sp.symbols('f g h', cls=sp.Function)
sp.init_printing(use_latex=True)
plt.rc('savefig', dpi=100)
plt.rc('savefig', format='png')
mp.mp.dps = 1000
MPF_ONE = mp.mpf('1')
#%load_ext sympy.interactive.ipythonprinting
```

```python
data = np.loadtxt('weib_diff_fit.txt', delimiter=';')
```

Location analysis
-----------------
* X - shape
* Y - number of filaments
* Z - location of the fitted Weibull distribution

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = data[:, 0].reshape(18, 59)
Y = data[:, 1].reshape(18, 59)
LOC = data[:, 2].reshape(18, 59)
ax.plot_wireframe(X, Y, LOC, rstride=1, cstride=1)
ax.plot_wireframe(X, Y, np.zeros_like(LOC), rstride=1, cstride=1, color='green')
ax.set_xlabel('shape')
ax.set_ylabel('number of filaments')
ax.set_zlabel('weibull location')
```

loglogplot location/ number of filaments/ shape versus number of filaments

```python
plt.figure()
plt.loglog(Y.T, (LOC / X).T)
```

plot modified differences between above plotted lines

```python
A = X.T[-1, :]
B = LOC.T[-1, :]
ONES = np.ones_like(A)
x=np.linspace(2,100,98)

plt.loglog(A,B)
plt.loglog(x,2.*x**(-0.8))
```

scale analysis
--------------
* X - shape
* Y - number of filaments
* Z - scale of the fitted Weibull distribution

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
SCALE = data[:, 3].reshape(18, 59)
ax.plot_wireframe(X, Y, SCALE, rstride=2, cstride=2)
ax.plot_wireframe(X, Y, np.ones_like(SCALE), rstride=1, cstride=1, color='green')
ax.set_xlabel('shape')
ax.set_ylabel('number of filaments')
ax.set_zlabel('weibull scale')
```

```python
plt.figure()
plt.plot(Y.T, SCALE.T)
#plt.plot(Y.T[:, 0], 1.0 / Y.T[:, 0], 'k-', linewidth=2)
```

```python
A = X.T[-1, :]
B = SCALE.T[-1, :]
ONES = np.ones_like(A)
plt.plot(A,B)
```

shape analysis
--------------
* X - shape
* Y - number of filaments
* Z - shape of the fitted Weibull distribution

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
SHAPE = data[:, 4].reshape(18, 59)
ax.plot_wireframe(X, Y, np.log(SHAPE), rstride=2, cstride=2)
ax.plot_wireframe(X, Y, np.log(np.ones_like(SHAPE)), rstride=1, cstride=1, color='green')
ax.set_xlabel('shape')
ax.set_ylabel('number of filaments')
ax.set_zlabel('weibull shape')
```

```python
plt.figure()
plt.plot(Y.T, SHAPE.T)
#plt.plot(Y.T[:, 0], 1.0 / Y.T[:, 0], 'k-', linewidth=2)
```

```python
A = X.T[-1, :]
B = SHAPE.T[-1, :]
ONES = np.ones_like(A)
plt.plot(A,B)
```
