```python
from IPython.display import display, HTML
import os
import numpy as np
import mpmath as mp
import matplotlib.pyplot as plt
import sympy as sp
#sp.init_session()
#x, y, z, t = sp.symbols('x y z t')
#k, m, n = sp.symbols('k m n', integer=True)
#f, g, h = sp.symbols('f g h', cls=sp.Function)
sp.init_printing(use_latex=True)
plt.rc('savefig', dpi=100)
plt.rc('savefig', format='png')
mp.mp.dps = 1000
MPF_ONE = mp.mpf('1')
#%load_ext sympy.interactive.ipythonprinting
```

```python
w_arr = np.array([0, 0.001812, 0.007698, 0.012549, 0.017968, 0.025943, 0.040328, 0.069539, 0.123633, 0.3])/1000.
ft_arr = np.array([2.7, 2.7, 1.697259, 1.176705, 0.884239, 0.672137, 0.465486, 0.254721, 0.100837, 0])
Gf = np.sum((ft_arr[:-1] + ft_arr[1:]) * (w_arr[1:] - w_arr[:-1]) / 2.)
ft = ft_arr[0]
print Gf
plt.plot(w_arr,ft_arr)
```

Hordijk
-------

```python
wc = 5.14 * Gf / ft
c1 = 3.
c2 = 6.39
w_hordijk = np.copy(w_arr)
w_hordijk[-1] = wc
ft_hordijk = ft * ((1 + (c1 * w_hordijk/wc)**3) * np.exp(-c2 * w_hordijk/wc) - w_hordijk/wc * (1 + c1**3) * np.exp(-c2))
plt.plot(w_hordijk, ft_hordijk)
plt.plot(w_arr, ft_arr)
```

Linear softening
----------------

```python
wf = 2 * Gf / ft
w_linear = np.array([0, wf])
ft_linear = ft * (1 - w_linear / wf)
plt.plot(w_linear, ft_linear)
plt.plot(w_arr, ft_arr)
```

Bilinear
--------

```python
par = 2.0 # 2.5
Gf1 = Gf/par
Gf2 = Gf-Gf1
wf1 = 2 * Gf1 / ft

dwf = 0.0003 - wf1
dft = 2 * Gf2 / dwf

w_bilinear = np.array([0, wf1])
ft_bilinear = ft * (1 - w_bilinear / wf1)

w_bilinear = np.append(w_bilinear, wf1)
ft_bilinear = np.append(ft_bilinear, dft)
w_bilinear = np.append(w_bilinear, 0.0003)
ft_bilinear = np.append(ft_bilinear, 0.0)
plt.plot(w_bilinear, ft_bilinear)
plt.plot(w_arr, ft_arr)
```

exponential
-----------

```python
wf = Gf/ft
w_exp = np.linspace(0, 0.0003,20)
ft_exp = ft * np.exp(-w_exp/wf)
plt.plot(w_exp, ft_exp)
plt.plot(w_arr, ft_arr)
```
