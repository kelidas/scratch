```python
data_old = np.load(os.path.join(path,'t_old.npa'))
data_old_mp = np.load(os.path.join(path,'t_old_mp.npa'))
data_old_mp_gmpy = np.load(os.path.join(path,'t_old_mp_gmpy.npa'))
data_old_c = np.load(os.path.join(path,'t_old_c.npa'))
data_100 = np.load(os.path.join(path,'tmp_100.npa'))
data_1000 = np.load(os.path.join(path,'tmp_1000.npa'))
data_1000_gmpy = np.load(os.path.join(path,'tmp_1000_gmpy.npa'))
data_10000 = np.load(os.path.join(path,'tmp_10000.npa'))
```

Old loop version
----------------
* old -- double precision
* old_mp -- mpmath precision set to 1000
* old_mp_gmpy -- mpmath precision set to 1000 and BACKAND="gmpy"
* old_c -- recursion implemented in c (prec 1000)

old_c version is approximatelly 2 times faster then old_mp

```python
plt.plot(data_old[:, 0], data_old[:, 1], label='old')
plt.plot(data_old_mp[:, 0], data_old_mp[:, 1], label='old_mp')
plt.plot(data_old_mp_gmpy[:, 0], data_old_mp_gmpy[:, 1], label='old_mp_gmpy')
plt.plot(data_old_c[:, 0], data_old_c[:, 1], label='old_c')
plt.legend(loc='best')
```

```python
plt.semilogy(data_old[:, 0], data_old[:, 1], label='old')
plt.semilogy(data_old_mp[:, 0], data_old_mp[:, 1], label='old_mp')
plt.semilogy(data_old_mp_gmpy[:, 0], data_old_mp_gmpy[:, 1], label='old_mp_gmpy')
plt.semilogy(data_old_c[:, 0], data_old_c[:, 1], label='old_c')
plt.legend(loc='best')
```

New MP version

```python
plt.plot(data_100[:, 0], data_100[:, 1], label='100')

plt.plot(data_1000[:, 0], data_1000[:, 1], label='1000')
plt.plot(data_1000_gmpy[:, 0], data_1000_gmpy[:, 1], label='1000_gmpy')

plt.plot(data_10000[:, 0], data_10000[:, 1], label='10000')
plt.legend(loc='best')
```
